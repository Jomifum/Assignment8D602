### **Part 1: Matplotlib Visualizations**

```
#Import Libraries
import matplotlib.pyplot as plt

#Scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(data['Shape__Area'], data['Obesity'], color='blue', label='Obesity Rate')
plt.title('Obesity Rate vs. Shape Area')
plt.xlabel('Shape Area')
plt.ylabel('Obesity Rate')
plt.legend(loc='upper right', fontsize=12)
plt.show()

#Bar plot
plt.figure(figsize=(12, 8))
plt.bar(data['NAME'], data['Obesity'], color='green')
plt.title('Obesity Rates by State')
plt.xlabel('State')
plt.ylabel('Obesity Rate')
plt.xticks(rotation=90)
plt.show()
```

##Part 2: Seaborn Visualizations

```
#Import Libraries
import seaborn as sns

#Scatter plot:
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Shape__Area', y='Obesity', data=data, hue='Obesity', palette='viridis')
plt.title('Obesity Rate vs. Shape Area')
plt.xlabel('Shape Area')
plt.ylabel('Obesity Rate')
plt.legend(loc='upper right')
plt.show()

#Bar plot:
plt.figure(figsize=(12, 8))
sns.barplot(x='NAME', y='Obesity', data=data, palette='magma')
plt.title('Obesity Rates by State')
plt.xlabel('State')
plt.ylabel('Obesity Rate')
plt.xticks(rotation=90)
plt.show()
```

##Differences Between Matplotlib and Seaborn
Matplotlib and Seaborn are both powerful visualization libraries in Python, but they have different strengths:

Matplotlib: Offers more control over the plot and is highly customizable. It requires more code to create complex visualizations.

Seaborn: Built on top of Matplotlib, it provides a high-level interface for drawing attractive and informative statistical graphics with less code. It is easier to create complex visualizations quickly.

##Visualization Top 10 states with Obesity

```
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
url = "https://raw.githubusercontent.com/Jomifum/Assignment8D602/main/LakeCounty_Health_-6177935595181947989.csv"
data = pd.read_csv(url)

# Sort the dataset by obesity rates in descending order and select the top 10 states
top_10_obesity = data.nlargest(10, 'Obesity')

# Create the bar plot
plt.figure(figsize=(12, 8))
plt.bar(top_10_obesity['NAME'], top_10_obesity['Obesity'], color='teal')
plt.title('Top 10 States with Highest Obesity Rates')
plt.xlabel('State')
plt.ylabel('Obesity Rate (%)')
plt.xticks(rotation=45)
plt.show()
```

##Interpretation:
The visualization highlights the top 10 states with the highest obesity rates. This provides a clear understanding of which states have the most significant obesity challenges, which can be crucial for targeted public health interventions and resource allocation.
